# Gap Gini Score Example

This notebook demonstrates how to compute **Gap Gini scores** for pairwise sequence alignments.

The Gap Gini score measures how uniformly gaps are distributed in a pairwise alignment. A low score (< 0.1) indicates evenly distributed gaps (good alignment), while a high score indicates concentrated gaps — a sign of truncation or large indels.

## Imports

```python
from bio_programming_tools.tools.sequence_alignment.gap_gini import (
    run_gap_gini, GapGiniInput, GapGiniConfig,
)
```

## 1. Score a Pairwise Alignment

Compute the Gap Gini score for a FASTA-formatted pairwise alignment. Each alignment must contain exactly 2 sequences.

### API Reference

**`GapGiniInput`**

| Field | Type | Description |
|-------|------|-------------|
| `alignments` | `List[str]` | FASTA-formatted pairwise alignments (2 sequences each) |

**`GapGiniConfig`**

No configuration parameters needed.

**`GapGiniOutput`**

| Field | Type | Description |
|-------|------|-------------|
| `gini_scores` | `List[float]` | Per-alignment Gini scores (0.0-1.0) |

```python
# Good alignment: gaps evenly distributed (low Gini)
good_alignment = ">reference\nMVLSPADKTNVKAAWGKVGA-HAGEYGAEALERMFLSFPTTK\n>candidate\nMVLSPADKTNVKAAWGKVGA-HAGEYGAEALERMFLSFPTTK"

# Bad alignment: gaps concentrated at one end (high Gini = truncation)
bad_alignment = ">reference\nMVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTKTYFPHFDLSH\n>candidate\nMVLSPADKTNVKAAWGKVGA-------------------------------"

inputs = GapGiniInput(alignments=[good_alignment, bad_alignment])
config = GapGiniConfig()

result = run_gap_gini(inputs, config)

GAP_GINI_THRESHOLD = 0.1
labels = ["Good (distributed gaps)", "Bad (concentrated gaps)"]
for label, score in zip(labels, result.gini_scores):
    status = "PASS" if score < GAP_GINI_THRESHOLD else "FAIL (truncated)"
    print(f"{label}: gini={score:.3f} -> {status}")
```

## 2. Export Results

Save Gini scores to CSV or JSON.

```python
from pathlib import Path

output_dir = Path("./example_output")
result.export(name="gap_gini_scores", export_path=output_dir, file_format="json")
print(f"Gap Gini scores exported to {output_dir}")
```
