# 🧬 ViennaRNA Tool Example

This notebook demonstrates how to predict RNA secondary structure using **ViennaRNA**.

## 📖 What is ViennaRNA?

[ViennaRNA](https://www.tbi.univie.ac.at/RNA/) is a comprehensive package for RNA secondary structure prediction and comparison. It uses thermodynamic models to compute minimum free energy (MFE) structures and partition functions for RNA sequences.

### Key Features:
- **Structure prediction** -- Compute minimum free energy (MFE) secondary structures
- **Thermodynamic modeling** -- Energy calculations based on Turner or Mathews parameters
- **DNA & RNA support** -- Fold both RNA and DNA sequences with appropriate energy parameters
- **Flexible control** -- Temperature, base pair constraints, and parameter set options

## 📥 Imports

```python
from bio_programming_tools import ViennaRNAConfig, ViennaRNAInput, run_viennarna
```

## 🔍 1. Predict RNA Secondary Structure

Predict the minimum free energy (MFE) secondary structure for RNA sequences.

ViennaRNA computes the thermodynamically most stable folding for each sequence, returning the dot-bracket structure notation and the free energy in kcal/mol.

### API Reference

**`ViennaRNAInput`**

| Field | Type | Description |
|-------|------|-------------|
| `sequences` | `List[str]` | RNA sequences to fold. Valid nucleotides: `A`, `U`, `G`, `C`, `T`, `N` (case-insensitive). DNA sequences containing `T` are converted to RNA |

**`ViennaRNAConfig`**

| Field | Type | Default | Description |
|-------|------|---------|-------------|
| `temperature` | `float` | `37.0` | Temperature in Celsius for energy calculations (min: `-273.15`) |
| `use_dna_params` | `bool` | `False` | Use DNA energy parameters instead of RNA parameters |
| `no_lonely_pairs` | `bool` | `False` | Disallow lonely base pairs (helices of length 1) |

**`ViennaRNAOutput`**

| Field | Type | Description |
|-------|------|-------------|
| `results` | `List[ViennaRNAResult]` | Fold results, one per input sequence |

**`ViennaRNAResult`**

| Field | Type | Description |
|-------|------|-------------|
| `sequence` | `str` | The input RNA sequence |
| `structure` | `Optional[str]` | Predicted secondary structure in dot-bracket notation |
| `mfe` | `Optional[float]` | Minimum free energy in kcal/mol |

```python
inputs = ViennaRNAInput(sequences=[
    "GCGCUUUUGCGC",
    "GGGGAAACCCC",
    "CGCGAAACGCG",
    "UUUUAAAAAUUUUU",
    "GCUAGCUAGC"
])

# Configure folding parameters
config = ViennaRNAConfig(
    temperature=37.0  # Physiological temperature in Celsius
)

result = run_viennarna(inputs, config)

# Display the predicted structures
for i, res in enumerate(result.results):
    print(f"Sequence {i+1}: {res.sequence}")
    print(f"  Structure: {res.structure}")
    print(f"  MFE: {res.mfe:.2f} kcal/mol")
    print()
```

## 💾 2. Export Results

Save the folding results to files for downstream analysis.

### Supported formats:
- **JSON** -- Structured data with metadata and all folding information
- **CSV** -- Tabular format with sequence, structure, and MFE columns

The exported results can be used for:
- Comparing folding energies across sequence variants
- Input to RNA design or optimization pipelines
- Visualization of secondary structures

```python
from pathlib import Path

# Create output directory
output_dir = Path("./example_output")
output_dir.mkdir(exist_ok=True)

# Export results to CSV format
result.export(output_dir / "viennarna_results.csv", file_format="csv")
print("Results exported to ./example_output/")
```
