# CCD Database Explorer

This notebook explores the Chemical Component Dictionary (CCD) database from wwPDB.

The database contains SMILES representations, CCD codes, and descriptions for chemical components found in PDB structures.

```python
import pandas as pd
from bio_programming.bio_tools.entities.ligands import CCD_DATABASE_PATH
```

## Load CCD Database into DataFrame

```python
# Read the tab-delimited file
# Columns: SMILES, CCD_Code, Description
ccd_df = pd.read_csv(
    CCD_DATABASE_PATH,
    sep='\t',
    header=None,
    names=['smiles', 'ccd_code', 'description']
)

print(f"Loaded {len(ccd_df):,} entries from CCD database")
ccd_df.head()
```

## Database Statistics

```python
print(f"Total entries: {len(ccd_df):,}")
print(f"Unique CCD codes: {ccd_df['ccd_code'].nunique():,}")
print(f"Unique SMILES: {ccd_df['smiles'].nunique():,}")
print(f"\nMissing values:")
print(ccd_df.isnull().sum())
```

# CCD lookup tests

```python
from bio_programming.bio_tools.entities.ligands import map_smiles_to_ccd_code

# Test SMILES from the failing test and other examples
test_smiles = [
    "CNC[C@@H](c1ccc(c(c1)O)O)O",  # L-epinephrine
    "c1nc(c2c(n1)n(cn2)[C@H]3[C@@H]([C@@H]([C@H](O3)CO[P@@](=O)(O)O[P@](=O)(O)OP(=O)(O)O)O)O)N",  # ATP

    # Add more SMILES to test here
    "CCO",  # Ethanol
    "C1=CC=CC=C1",  # Benzene

    "c1cc(ccc1C[C@@H](C(=O)O)N)O"
]

results = []
for smiles in test_smiles:
    try:
        ccd_code = map_smiles_to_ccd_code(smiles)
        results.append({
            "smiles": smiles,
            "ccd_code": ccd_code,
            "status": "✓ Found"
        })
        print(f"✓ {smiles[:50]:50s} -> {ccd_code}")
    except ValueError as e:
        results.append({
            "smiles": smiles,
            "ccd_code": None,
            "status": "✗ Not found",
            "error": str(e)
        })
        print(f"✗ {smiles[:50]:50s} -> NOT FOUND")
    except Exception as e:
        results.append({
            "smiles": smiles,
            "ccd_code": None,
            "status": "✗ Error",
            "error": str(e)
        })
        print(f"✗ {smiles[:50]:50s} -> ERROR: {e}")

# Summary
print(f"\n{'='*80}")
print(f"Summary: {sum(1 for r in results if r['status'] == '✓ Found')}/{len(results)} SMILES found in CCD database")
print(f"{'='*80}")
```
