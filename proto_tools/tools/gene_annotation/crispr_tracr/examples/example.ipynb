# CRISPRtracrRNA Prediction Example

This notebook demonstrates how to predict tracrRNA sequences from nucleotide CRISPR loci using **CRISPRtracrRNA**.

tracrRNAs are essential components of Type II CRISPR-Cas systems. This tool identifies candidate tracrRNA sequences by analyzing anti-repeat similarity and RNA-RNA interaction energies.

## Imports

```python
from proto_tools import CrisprTracrConfig, CrisprTracrInput, run_crispr_tracr
```

## 1. Predict tracrRNA

Scan a CRISPR locus for tracrRNA candidates. The tool analyzes anti-repeat complementarity and RNA interaction energies to locate the tracrRNA.

### API Reference

**`CrisprTracrInput`**

| Field | Type | Description |
|-------|------|-------------|
| `sequences` | `List[str]` | Nucleotide sequences containing CRISPR loci |
| `sequence_ids` | `Optional[List[str]]` | Identifiers for each sequence (auto-generated if omitted) |

**`CrisprTracrConfig`**

| Field | Type | Default | Description |
|-------|------|---------|-------------|
| `model_type` | `Literal["II", "all"]` | `"II"` | CRISPR system type to search for |
| `run_type` | `str` | `"default"` | Execution mode |
| `num_workers` | `Optional[int]` | `None` | Number of parallel workers |

**`CrisprTracrOutput`**

| Field | Type | Description |
|-------|------|-------------|
| `predictions` | `List[TracrPrediction]` | Per-sequence tracrRNA predictions |
| `metadata` | `Dict` | Execution metadata |

**`TracrPrediction`**

| Field | Type | Description |
|-------|------|-------------|
| `sequence_id` | `str` | Input sequence identifier |
| `tracr_start` | `Optional[int]` | tracrRNA start position (None if not found) |
| `tracr_end` | `Optional[int]` | tracrRNA end position (None if not found) |
| `interaction_energy` | `Optional[float]` | RNA-RNA interaction energy (kcal/mol) |
| `has_tracr` | `bool` | Whether a tracrRNA was detected (property) |

```python
# Provide a genomic sequence containing a CRISPR locus
# In practice, use a real CRISPR-containing genomic region
inputs = CrisprTracrInput(
    sequences=["ATGCGTAAACGATTGCAGT" * 500],  # Placeholder — use real CRISPR locus DNA
    sequence_ids=["locus_1"],
)

config = CrisprTracrConfig(model_type="II")

result = run_crispr_tracr(inputs, config)

for pred in result.predictions:
    if pred.has_tracr:
        print(f"{pred.sequence_id}: tracrRNA at {pred.tracr_start}-{pred.tracr_end}")
        print(f"  Interaction energy: {pred.interaction_energy} kcal/mol")
    else:
        print(f"{pred.sequence_id}: no tracrRNA detected")
```

## 2. Export Results

Save tracrRNA predictions for downstream analysis.

```python
from pathlib import Path

output_dir = Path("./example_output")
result.export(name="crispr_tracr_results", export_path=output_dir, file_format="json")
print(f"tracrRNA predictions exported to {output_dir}")
```
