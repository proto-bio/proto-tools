# DeepPBS Specificity

DeepPBS predicts the DNA base preferences of a protein directly from a protein-DNA complex structure. Given one or more PDB files, the `deeppbs-specificity` tool returns a canonical DNA-only position probability matrix (PPM) in `A,C,G,T` order, alongside the true DNA sequence, residue masks, and per-base chain labels.

This tool requires a local DeepPBS repository and a local X3DNA install. When a dependency is missing or processing fails, the tool returns a uniform fallback PPM flagged with `used_fallback`.

## Available tools

```python
from proto_tools.utils.notebook_docs import display_available_tools

display_available_tools("deeppbs_specificity")
```

## Usage

Provide one or more protein-DNA PDB structures. Each result carries a `predicted_ppm` (`L x 4`, ACGT order), the path to a canonical `.npz` artifact, and the `used_fallback` / `fallback_reason` markers.

```python
from proto_tools import (
    DeepPBSSpecificityConfig,
    DeepPBSSpecificityInput,
    run_deeppbs_specificity,
)

# Replace with a path to a real protein-DNA co-crystal PDB on your system.
pdb_path = "complex.pdb"

inputs = DeepPBSSpecificityInput(pdb_paths=[pdb_path])
config = DeepPBSSpecificityConfig()

output = run_deeppbs_specificity(inputs, config)

result = output.results[0]
print("used_fallback:", result.used_fallback)
print("fallback_reason:", result.fallback_reason)
print("PPM shape:", len(result.predicted_ppm), "x", len(result.predicted_ppm[0]))
print("output_npz_path:", result.output_npz_path)
```
