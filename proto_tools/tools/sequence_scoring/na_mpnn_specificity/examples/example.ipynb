# NA-MPNN Specificity

[NA-MPNN](https://github.com/akubaney/NA-MPNN) predicts protein–DNA specificity from a protein–DNA complex structure. The `na-mpnn-specificity` tool returns a canonical DNA-only position probability matrix (PPM) in `A,C,G,T` order for each input structure, along with the recovered true sequence, per-row masks, chain labels, and a path to a canonical `.npz` artifact.

Preprint: [Kubaney et al., 2025](https://www.biorxiv.org/content/10.1101/2025.10.03.679414v2)

> NA-MPNN requires a local repository checkout and specificity checkpoint on disk; neither is auto-downloaded. Set `na_mpnn_repo_path` and `checkpoint_path` accordingly, and run on a GPU host.

```python
from proto_tools.tools.sequence_scoring.na_mpnn_specificity import (
    NAMPNNSpecificityConfig,
    NAMPNNSpecificityInput,
    run_na_mpnn_specificity,
)
from proto_tools.utils.notebook_docs import display_api_reference
```

## Basic Usage

Score a protein–DNA complex for DNA specificity. Replace the path with a real protein–DNA complex PDB on your machine.

```python
inputs = NAMPNNSpecificityInput(pdb_paths=["protein_dna_complex.pdb"])
# Provide the repo + checkpoint via config below, or set $NA_MPNN_REPO_PATH / $NA_MPNN_CHECKPOINT_PATH.
config = NAMPNNSpecificityConfig()

result = run_na_mpnn_specificity(inputs, config)

first = result.results[0]
print("input:", first.input_name)
print("PPM shape:", len(first.predicted_ppm), "x", len(first.predicted_ppm[0]))
print("first PPM row (A,C,G,T):", first.predicted_ppm[0])
print("chain labels:", first.chain_labels)
print("npz path:", first.output_npz_path)
```

## Advanced Usage

Override the repository/checkpoint paths, sampling temperature, and persist canonical `.npz` artifacts to a chosen directory.

```python
config = NAMPNNSpecificityConfig(
    na_mpnn_repo_path="/path/to/NA-MPNN",
    checkpoint_path="/path/to/NA-MPNN/models/specificity_model/s_70114.pt",
    temperature=0.2,
    number_of_batches=2,
    output_directory="./na_mpnn_results",
    keep_intermediate=True,
)

result = run_na_mpnn_specificity(
    NAMPNNSpecificityInput(pdb_paths=["protein_dna_complex.pdb"]),
    config,
)
print("num results:", len(result.results))
```

## Export

Export the canonical results to JSON or CSV.

```python
result.export("na_mpnn_specificity_results", file_format="json")
print("Exported to na_mpnn_specificity_results.json")
```
